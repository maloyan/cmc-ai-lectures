![](images/mse.png)

Гипотеза (модель): $ h_\theta(x) = \theta_0 + \theta_1 \cdot x $

Параметры: $ \theta_0 $, $ \theta_1 $

Функция потерь: $ J(\theta_0, \theta_1) = \frac{1}{2m} \sum_{i=1}^m (h_\theta(x_i) - y_i)^2 $

Цель: минимизировать $ J(\theta_0, \theta_1) $

Производные:
1. **Для $ \theta_0 $**:
   $ \frac{\partial J}{\partial \theta_0} = \frac{1}{m} \sum_{i=1}^m (\theta_0 + \theta_1 x_i - y_i) $

2. **Для $ \theta_1 $**:
   $ \frac{\partial J}{\partial \theta_1} = \frac{1}{m} \sum_{i=1}^m (\theta_0 + \theta_1 x_i - y_i) x_i $

Решение:
1. $ \sum_{i=1}^m (\theta_0 + \theta_1 x_i - y_i) = 0 $
2. $ \sum_{i=1}^m (\theta_0 + \theta_1 x_i - y_i) x_i = 0 $

$ \theta_1 = \frac{\sum_{i=1}^{m} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{m} (x_i - \bar{x})^2} $

$ \theta_0 = \bar{y} - \theta_1 \bar{x} $

Где $ \bar{x} $ - это среднее значение $ x $, а $ \bar{y} $ - среднее значение $ y $.

```python
def mse(theta0, theta1, X, Y):
    m = len(X)
    sq_diff = [(hypothesis(theta0, theta1, x) - y)**2 for x, y in zip(X, Y)]
    return 0.5 / m * np.sum(sq_diff)
```

```python
from matplotlib import cm
from matplotlib.ticker import LinearLocator

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
THETA0 = np.arange(-100, 156, 10)
THETA1 = np.arange(-100, 156, 10)
THETA0, THETA1 = np.meshgrid(THETA0, THETA1)

Z = np.array([[mse(theta0, theta1, df.x.values, df.y.values) for theta1 in THETA1] for theta0 in THETA0])
# Plot the surface.
surf = ax.plot_surface(THETA0, THETA1, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(Z.min(), Z.max())
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

### Алгоритм градиентного спуска

Градиентный спуск — это итерационный алгоритм оптимизации, используемый для нахождения минимума функции. Основная идея состоит в инициализации параметров функции, а затем в их итеративной корректировке для минимизации целевой функции.

#### Формула:
Правило обновления в алгоритме градиентного спуска для параметра $ \theta $ задаётся следующим образом:

$ \theta := \theta - \alpha \frac{\partial}{\partial \theta} J(\Theta) $

Где:
- $ \theta $ — параметр функции, минимум которой мы пытаемся найти.
- $ \alpha $ — скорость обучения.
- $ \frac{\partial}{\partial \theta} J(\Theta) $ — частная производная целевой функции $ J(\Theta) $ по параметру $ \theta $.

#### Шаги:
1. **Инициализация**: Начните с начальных предположений для параметров. Эти значения могут быть случайными или инициализированными нулями.
2. **Обновление**: Для каждого параметра $ \theta $ обновите его **одновременно**, используя указанное выше правило обновления.
3. **Итерация**: Повторяйте шаг обновления до сходимости параметров. Это означает, что изменение целевой функции становится незначительным или достигнуто предопределенное количество итераций.

#### Примечания:
- Скорость обучения $ \alpha $ требует тщательной настройки. Слишком маленькое значение приводит к медленной сходимости, в то время как слишком большое может вызвать проскакивание минимума или невозможность сходимости.
- Необходимо определить критерии сходимости, например, порог для изменения целевой функции или максимальное количество итераций.
- Градиентный спуск может использоваться для широкого круга задач, включая нелинейную регрессию, классификацию и более сложные алгоритмы машинного обучения.


```python
import matplotlib.animation as animation
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Импортирование данных из CSV-файла в DataFrame Pandas
df_test = pd.read_csv("train.csv")

print("----------Test Data----------")
print(df_test.head())  # Вывод первых пяти строк DataFrame

# Преобразование данных в массив для последующих расчетов
x_test = np.array(df_test["x"])
y_test = np.array(df_test["y"])

n = len(x_test)  # Количество образцов в данных

iter = 0  # Начальный счетчик итераций
itersteps = 250  # Общее количество итераций

# Начальные предположения для линейной регрессионной модели y=theta1*x+theta0 и массивы для отслеживания значений в течение итераций
theta1 = np.zeros(itersteps)
theta0 = np.zeros(itersteps)
J = np.zeros(itersteps)  # Среднеквадратичная ошибка
dJdtheta1 = np.zeros(itersteps)
dJdtheta0 = np.zeros(itersteps)

learning_rate = 1e-5  # Скорость обучения

for iter in range(itersteps - 1):

    y_new = theta1[iter] * x_test + theta0[iter]

    J[iter] = 1 / (2 * n) * np.sum((y_test - y_new) ** 2)  # Среднеквадратичная ошибка

    dJdtheta0[iter] = 1 / n * np.sum((y_new - y_test))
    dJdtheta1[iter] = 1 / n * np.sum((y_new - y_test) * x_test)
    
    theta1[iter + 1] = theta1[iter] - learning_rate * dJdtheta1[iter]
    theta0[iter + 1] = theta0[iter] - learning_rate * dJdtheta0[iter]

# Данные для создания гладкой линии линейной регрессионной модели
x_min = min(x_test)
x_max = max(x_test)
x = np.linspace(x_min, x_max, n)
y = theta1[0] * x + theta0[0]

# Создание начальной фигуры
fig1, ax = plt.subplots(figsize=(10, 10))
ax.scatter(x_test, y_test)  # Начальный диаграмма рассеяния
(line,) = ax.plot(x, y, "red")  # Начальная линейная подгонка
ax.set_xlabel("X")
ax.set_ylabel("Y")
t1 = ax.text(
    50, 25, f"Eqn: y={str(round(theta1[iter], 4))}*x+{str(round(theta0[iter], 4))}", fontsize=15
)

# Функция обновляет данные для анимации графика с помощью animation.FuncAnimation() ниже
# Переменная, которая передается этой функции из FuncAnimation, это frames=itersteps-1
# Это позволяет получить данные на каждом шаге итерации
def update(iter):
    y = theta1[iter] * x + theta0[iter]  # Данные линейной подгонки на каждой итерации
    line.set_data(x, y)  # Обновление данных линейной подгонки для графика
    t1.set_text(f"Eqn: y={str(round(theta1[iter], 4))}*x+{str(round(theta0[iter], 4))}")
    ax.set_title(f"Linear Regression MSE: {J[iter]} iteration: {str(iter)}")
    return line, ax

# Функция анимации после расчета всех переменных на каждой итерации
# Вызывает функцию update и передает количество кадров=itersteps-1, чтобы получить все данные на каждой итерации
ani = animation.FuncAnimation(
    fig1, update, frames=itersteps - 1, interval=10, blit=False, repeat_delay=100
)
ani.save("lin_reg.mp4", fps=30, extra_args=["-vcodec", "libx264"])  # Сохранение анимации

fig2, ax2 = plt.subplots(figsize=(10, 10))

ax2.plot(J, label="SSE", color="blue")

ax2.set_title("Errors")
ax2.set_xlabel("Number of Iterations")
ax2.set_ylabel("Error", color="blue")
ax2.tick_params("y", colors="blue")

fig2.savefig("errors.png")  # Сохранение рисунка
```

Нормализация данных — это процесс масштабирования входных переменных таким образом, чтобы они имели общий масштаб. Этот процесс важен по нескольким причинам, особенно в области машинного обучения и оптимизации:

### 1. **Улучшение сходимости:**
   - **Более быстрая оптимизация:** Когда все признаки имеют разный масштаб, градиенты могут быть очень разными и могут потребовать различных скоростей обучения. Нормализация помогает унифицировать масштабы, что ускоряет сходимость алгоритма.
   - **Предотвращение доминирования признаков:** Без нормализации признаки с большими значениями могут доминировать в процессе обучения, влияя на производительность модели.

### 2. **Стабильность численных вычислений:**
   - **Избегание численных проблем:** Большие значения могут вызвать численную нестабильность в некоторых вычислениях, таких как вычисление градиентов. Нормализация помогает избежать таких проблем.

### 3. **Обучение с более высокой скоростью:**
   - **Эффективное использование градиентного спуска:** Градиентный спуск и другие методы оптимизации работают лучше, когда все признаки находятся на схожем масштабе.

### 4. **Облегчение выбора гиперпараметров:**
   - **Упрощение настройки параметров:** Нормализация делает процесс выбора параметров, таких как скорость обучения, более предсказуемым и менее зависимым от масштаба признаков.

### Типы нормализации:
1. **Min-Max нормализация:** Приводит все значения к диапазону [0, 1] или другому заданному диапазону.
2. **Z-оценка или стандартизация:** Приводит данные к форме, где среднее значение равно 0, а стандартное отклонение — 1.

## Практические советы

### Отрицательные значения функции потерь
1. **Ошибки в реализации функции потерь:** Проверьте, правильно ли реализована и используется ваша функция потерь. Для многих стандартных функций потерь (например, MSE или кросс-энтропия) отрицательные значения не являются нормальными.
2. **Специфические функции потерь:** Некоторые функции потерь (например, логарифмические функции) могут теоретически давать отрицательные значения. Убедитесь, что вы полностью понимаете, как работает ваша функция потерь.

### Стагнирующие значения функции потерь
1. **Недостаточная скорость обучения:** Если скорость обучения слишком мала, процесс обучения может застрять на некотором уровне ошибки и не улучшаться.
2. **Недостаточно данных:** Модель может не иметь достаточно данных для обучения или данные могут быть не очень разнообразными.
3. **Слишком простая модель:** Модель может быть недостаточно сложной, чтобы выучить закономерности в данных.

## Логистическая регрессия

Логистическая регрессия — это статистический метод для анализа набора данных, в котором есть одна или несколько независимых переменных, определяющих исход. Несмотря на название, логистическая регрессия используется для **бинарной классификации**, а не для предсказания количественных значений. В отличие от линейной регрессии, которая использует линейное уравнение, логистическая регрессия использует логистическую функцию для моделирования бинарного исхода события.

### Основные характеристики
- **Выходное значение:** Вероятность того, что данный входной набор данных принадлежит к определенному классу.
- **Функция активации:** Логистическая (сигмоидная) функция, которая ограничивает выходное значение в диапазоне от 0 до 1.
- **Функция потерь:** Обычно используется функция потерь кросс-энтропии.

### Математическая модель
Если $ X $ - это входные переменные (признаки), то логистическая регрессия моделируется как:
$$ P(Y=1|X) = \sigma(\theta^T X) = \frac{1}{1 + e^{-\theta^T X}} $$
где:
- $ P(Y=1|X) $ - вероятность того, что класс равен 1 при данных значениях $ X $,
- $ \sigma $ - сигмоидная функция,
- $ \theta $ - параметры (веса) модели,
- $ \theta^T X $ - скалярное произведение векторов параметров и признаков.

```python
import numpy as np
import matplotlib.pyplot as plt

# Значения x от -10 до 10 с шагом 0.1
x = np.arange(-10., 10., 0.1)
# Сигмоидная функция
sigmoid = 1 / (1 + np.exp(-x))

# Визуализация сигмоиды
plt.figure(figsize=(10, 6))
plt.plot(x, sigmoid)
plt.title("Сигмоидная функция")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.grid(True)
plt.show
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn import datasets

# Загрузка датасета
iris = datasets.load_iris()
X = iris.data[:, :2]  # Используем только первые два признака
y = (iris.target != 0) * 1  # Бинарная классификация: Iris Setosa или не Setosa

# Разбиение на тренировочный и тестовый наборы данных
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Создание и обучение модели логистической регрессии
model = LogisticRegression()
model.fit(X_train, y_train)

# Предсказание и оценка модели
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
# Использование только одного признака (длины лепестка) для логистической регрессии
X_single_feature = iris.data[:, 2].reshape(-1, 1)  # Длина лепестка
y_single_feature = (iris.target != 0) * 1  # Бинарная классификация: Iris Setosa или не Setosa

# Разбиение данных на тренировочный и тестовый наборы
X_train_single, X_test_single, y_train_single, y_test_single = train_test_split(X_single_feature, y_single_feature, test_size=0.2, random_state=0)

# Обучение модели
model_single_feature = LogisticRegression()
model_single_feature.fit(X_train_single, y_train_single)

# Визуализация
plt.figure(figsize=(10, 6))

# Точки данных
plt.scatter(X_single_feature, y_single_feature, color='black', zorder=20)

# Генерация значений для сигмоиды
X_test_range = np.linspace(X_single_feature.min(), X_single_feature.max(), 300)
sigmoid = model_single_feature.predict_proba(X_test_range.reshape(-1, 1))[:, 1]

# Построение сигмоиды
plt.plot(X_test_range, sigmoid, color='red', linewidth=3)

# Настройка графика
plt.xlabel("Длина лепестка")
plt.ylabel("Вероятность принадлежности к классу 1")
plt.title("Визуализация логистической регрессии с использованием сигмоиды")
plt.axhline(y=0.5, color='blue', linestyle='--')  # Решающая граница
plt.ylim(-0.1, 1.1)
plt.xlim(X_single_feature.min() - 1, X_single_feature.max() + 1)

plt.show()
```

```python
# Визуализация логистической регрессии с использованием коэффициентов модели для отрисовки решающей границы

# Использование двух признаков (длины и ширины лепестка) для логистической регрессии
X_two_features = iris.data[:, :2]  # Первые два признака: длина и ширина лепестка
y_two_features = (iris.target != 0) * 1  # Бинарная классификация: Iris Setosa или не Setosa

# Разбиение данных на тренировочный и тестовый наборы
X_train_two, X_test_two, y_train_two, y_test_two = train_test_split(X_two_features, y_two_features, test_size=0.2, random_state=0)

# Обучение модели
model_two_features = LogisticRegression()
model_two_features.fit(X_train_two, y_train_two)

# Коэффициенты модели
coef = model_two_features.coef_[0]
intercept = model_two_features.intercept_

# Визуализация
plt.figure(figsize=(10, 6))

# Точки данных
plt.scatter(X_two_features[:, 0], X_two_features[:, 1], c=y_two_features, cmap=plt.cm.Paired, edgecolors='k')

# Генерация значений для решающей границы
x_values = np.array([X_two_features[:, 0].min(), X_two_features[:, 0].max()])
y_values = -(x_values * coef[0] + intercept) / coef[1]

# Построение решающей границы
plt.plot(x_values, y_values, label="Решающая граница", color='red')

# Настройка графика
plt.xlabel("Длина лепестка")
plt.ylabel("Ширина лепестка")
plt.title("Визуализация логистической регрессии с двумя признаками")
plt.legend()
plt.show()
```

На графике выше изображена решающая граница для логистической регрессии с двумя признаками: длиной ($ x_1 $) и шириной ($ x_2 $) лепестка ириса. 

Точки представляют наблюдаемые данные, где цвет указывает на класс: либо Iris Setosa (класс 0), либо не Setosa (класс 1). Красная линия показывает решающую границу, определяемую моделью логистической регрессии с параметрами $ \theta_0, \theta_1, \theta_2 $ (соответствуют `model.intercept_` и `model.coef_`).

Гипотеза $ h_\theta(x) $ в логистической регрессии задается как:
$$h_\theta(x) = g(\theta_0 + \theta_1 x_1 + \theta_2 x_2)$$
где $ g $ — сигмоидная функция.

Решающая граница определяется как место, где $ h_\theta(x) = 0.5 $, что соответствует условию:
$$\theta_0 + \theta_1 x_1 + \theta_2 x_2 = 0$$

Преобразуя это уравнение для отображения на графике, где $ x_1 $ и $ x_2 $ — координаты точек (длина и ширина лепестка соответственно), получаем:
$$x_2 = -\frac{\theta_0 + \theta_1 x_1}{\theta_2}$$

На графике значение $ x_1 $ (длина лепестка) откладывается по оси абсцисс, а $ x_2 $ (ширина лепестка) — по оси ординат.

## Нелинейная граница решений

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.datasets import make_circles

# 1. Генерация данных
X, y = make_circles(n_samples=400, factor=0.3, noise=0.1, random_state=0)

# 2. Построение модели
# Создаем полиномиальные признаки и логистическую регрессию
model = Pipeline([
    ('poly', PolynomialFeatures(degree=3)),
    ('logistic', LogisticRegression())
])
model.fit(X, y)

# 3. Визуализация
# Генерация сетки для визуализации решающей границы
x0_min, x0_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
x1_min, x1_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
xx0, xx1 = np.meshgrid(np.linspace(x0_min, x0_max, 100),
                       np.linspace(x1_min, x1_max, 100))

# Предсказания на сетке
Z = model.predict(np.c_[xx0.ravel(), xx1.ravel()])
Z = Z.reshape(xx0.shape)

# Визуализация
plt.figure(figsize=(10, 8))
plt.contourf(xx0, xx1, Z, alpha=0.8, cmap=plt.cm.bwr)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.bwr)
plt.title("Нелинейная граница решения (круг)")
plt.xlabel("Признак 1")
plt.ylabel("Признак 2")
plt.show()
```

### Исходная Задача
У нас есть набор данных $\{ (x_i, y_i) \}$, где $ x_i $ представляет признаки (в нашем случае два признака, соответствующие координатам на плоскости), а $ y_i $ — метки классов (0 или 1). Наша цель — найти функцию, которая сможет классифицировать эти точки, разделяя их границей в виде круга.

### Полиномиальные Признаки
Мы используем полиномиальные признаки для преобразования исходного пространства признаков таким образом, чтобы линейная модель смогла справиться с нелинейной задачей. Например, включая квадратичные признаки, мы преобразуем каждую точку данных $ x_i = (x_{i1}, x_{i2}) $ в новый набор признаков:

$$\phi(x_i) = (1, x_{i1}, x_{i2}, x_{i1}^2, x_{i1} x_{i2}, x_{i2}^2, \ldots)$$

### Логистическая Регрессия
Логистическая регрессия моделирует вероятность того, что данное наблюдение принадлежит к классу 1, с использованием сигмоидной функции:

$$h_\theta(x) = g(\theta^T \phi(x))$$

где $ g(z) = \frac{1}{1 + e^{-z}} $ — сигмоидная функция.

### Решающая Граница
Решающая граница определяется условием, при котором предсказанная моделью вероятность равна 0.5:

$$h_\theta(x) = 0.5 \quad \Rightarrow \quad \theta^T \phi(x) = 0$$

Для круговой границы, если мы возьмем только признаки до второй степени, решающая граница в исходном пространстве признаков будет определяться уравнением вида:

$$\theta_0 + \theta_1 x_1 + \theta_2 x_2 + \theta_3 x_1^2 + \theta_4 x_1 x_2 + \theta_5 x_2^2 = 0$$

Это уравнение представляет собой круг (или эллипс) в пространстве признаков $ (x_1, x_2) $. Таким образом, добавление полиномиальных признаков позволяет логистической регрессии формировать нелинейные решающие границы, такие как круги или эллипсы, в исходном пространстве.

## Функция потерь для логистической регрессии

Функция потерь (или функция стоимости) в логистической регрессии обычно определяется как бинарная кросс-энтропия или логистическая функция потерь. Эта функция измеряет, насколько предсказания модели отличаются от фактических меток. Для набора данных $\{ (x^{(i)}, y^{(i)}) \}_{i=1}^m$, где $ y^{(i)} $ принимает значение 0 или 1, функция потерь определяется следующим образом:

$$J(\theta) = -\frac{1}{m} \sum_{i=1}^{m} \left[ y^{(i)} \log(h_\theta(x^{(i)})) + (1 - y^{(i)}) \log(1 - h_\theta(x^{(i)})) \right]$$

где:
- $ m $ — количество примеров в наборе данных.
- $ h_\theta(x^{(i)}) $ — предсказанная моделью вероятность того, что $ x^{(i)} $ принадлежит классу 1. Она определяется как $ h_\theta(x) = g(\theta^T x) $, где $ g $ — сигмоидная функция, $ g(z) = \frac{1}{1 + e^{-z}} $.
- $ \theta $ — параметры модели (веса).

### Интуиция
Функция потерь логистической регрессии наказывает предсказания, которые далеки от фактических меток. Если фактическая метка $ y^{(i)} $ равна 1, то потеря увеличивается, когда $ h_\theta(x^{(i)}) $ стремится к 0. Если $ y^{(i)} $ равна 0, то потеря увеличивается, когда $ h_\theta(x^{(i)}) $ стремится к 1. Таким образом, эта функция потерь создает стимул для модели правильно предсказывать вероятность принадлежности каждого примера к классу 1.

```python
import numpy as np
import matplotlib.pyplot as plt

# Функция потерь для логистической регрессии для одного примера
def logistic_loss(y, h):
    return -(y * np.log(h) + (1 - y) * np.log(1 - h))

# Диапазон возможных предсказаний h от 0 до 1
h = np.linspace(0.01, 0.99, 100)

# Функция потерь для y = 1 и y = 0
loss_y1 = logistic_loss(1, h)
loss_y0 = logistic_loss(0, h)

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(h, loss_y1, label="Потери, если y=1", color="blue")
plt.plot(h, loss_y0, label="Потери, если y=0", color="red")
plt.title("Логистическая функция потерь")
plt.xlabel("h(x) - предсказанная вероятность")
plt.ylabel("Потери")
plt.legend()
plt.grid(True)
plt.show()
```

## Оптимизаторы

Методы оптимизации играют ключевую роль в обучении машинного обучения, помогая минимизировать (или максимизировать) функцию потерь или стоимости, адаптируя параметры модели. Рассмотрим несколько основных методов оптимизации и их характеристики.

### 1. Градиентный спуск (Gradient Descent)
**Формула:**
$$\theta = \theta - \alpha \nabla_\theta J(\theta)$$
**Описание:**
- $ \theta $ — параметры модели.
- $ \alpha $ — скорость обучения.
- $ \nabla_\theta J(\theta) $ — градиент функции потерь $ J $ по параметрам $ \theta $.
- **Вариации:** стохастический градиентный спуск (SGD), мини-batch градиентный спуск.

### 2. Стохастический градиентный спуск (Stochastic Gradient Descent, SGD)
**Формула:**
$$\theta = \theta - \alpha \nabla_\theta J(\theta; x^{(i)}; y^{(i)})$$
**Описание:**
- Вместо вычисления градиента на всём наборе данных, SGD делает это на каждом примере $ x^{(i)}, y^{(i)} $, что делает шаги более быстрыми, но менее точными.
- **Преимущества:** возможность избежать локальных минимумов, быстрее обрабатывает большие объемы данных.
- **Недостатки:** более высокая дисперсия обновлений, что приводит к нерегулярной сходимости.

### 3. Момент (Momentum)
**Формула:**
$$v_t = \gamma v_{t-1} + \alpha \nabla_\theta J(\theta)$$
$$\theta = \theta - v_t$$
**Описание:**
- Введение переменной $ v $ для ускорения SGD в нужном направлении, которая представляет собой комбинацию прошлых градиентов. Это как если бы мы накапливали "скорость" движения по поверхности функции потерь, учитывая предыдущие изменения.
- $ \gamma $ обычно установлен около 0.9.
- **Преимущества:** помогает ускорить SGD, особенно в направлении минимума и в случаях, где поверхность потерь имеет более крутой склон в одном направлении и более пологой в другом.

### 4. RMSprop (Root Mean Square Propagation)
**Формула:**
$$S_{dW} = \beta S_{dW} + (1 - \beta) (\nabla_{\theta} J(\theta))^2$$
$$\theta = \theta - \alpha \frac{\nabla_{\theta} J(\theta)}{\sqrt{S_{dW}} + \varepsilon}$$
**Описание:**
- Разработан для решения проблемы изменчивости шагов в SGD.
- $ S_{dW} $ — скользящее среднее квадрата градиентов.
- $ \varepsilon $ — маленькое число, предотвращающее деление на ноль.

### 5. Adam (Adaptive Moment Estimation)
**Формула:**
$$m_t = \beta_1 m_{t-1} + (1 - \beta_1) \nabla_{\theta} J(\theta)$$
$$v_t = \beta_2 v_{t-1} + (1 - \beta_2) (\nabla_{\theta} J(\theta))^2$$
$$\theta = \theta - \alpha \frac{m_t}{\sqrt{v_t} + \varepsilon}$$
**Описание:**
- Комбинирует идеи момента и RMSprop.
- $ m_t $ и $ v_t $ — оценки первого (среднее) и второго (невязанное среднее квадратичное) моментов градиентов.

## Decision Tree Classifier

**Decision Tree Classifier** — это модель машинного обучения, используемая для классификации или регрессии, которая работает по принципу дерева решений. Дерево решений представляет собой структуру, состоящую из узлов, ветвей и листьев. Каждый узел в дереве соответствует "вопросу" или "тесту" на атрибутах данных, ветви представляют возможные ответы на этот вопрос, а листья — конечные решения или классификации.
![](https://images.datacamp.com/image/upload/v1677504957/decision_tree_for_heart_attack_prevention_2140bd762d.png)


#### Как работает Decision Tree Classifier
1. **Выбор атрибута для разделения:** На каждом этапе дерево решений выбирает один из атрибутов и разделяет данные на основе этого атрибута. Этот процесс повторяется рекурсивно, пока не будет достигнут критерий останова (например, дерево достигло максимальной глубины или узлы содержат достаточно однородные данные).

2. **Критерии разделения:** Для выбора атрибута и разделения используются различные метрики, такие как Индекс Джини, Энтропия (в контексте усиления информации) и другие.

### Индекс Джини (Gini Impurity)
Индекс Джини — это мера, используемая для определения того, как хорошо разделение дерева разделяет данные на однородные классы. Индекс Джини равен 0, когда все элементы в узле принадлежат одному классу, и увеличивается до 0.5, когда классы распределены равномерно.

#### Формула Индекса Джини
$$\text{Gini}(D) = 1 - \sum_{i=1}^{n} p_i^2$$
где:
- $D$ — набор данных в узле.
- $p_i$ — вероятность того, что случайно выбранный элемент из набора данных $D$ будет классифицирован как класс $i$.
- $n$ — количество классов.

#### Интерпретация
- **Низкий Индекс Джини:** Узел содержит в основном объекты одного класса (близко к 0).
- **Высокий Индекс Джини:** Узел содержит смешанные объекты классов (близко к 0.5 для двух классов).

## Gini index (gini impurity)

- $N$ - Количество данных
- $p_i$ - Вероятность i-го класса

$Gini = \sum_{i=0}^{N-1} p_i(1-p_i)$

Для бинарного случая $p_0 = 1 - p_1$

$Gini = \sum_{i=0}^{N-1} p_i(1-p_i) = p_0(1-p_0) + p_1(1-p_1) = 2p_0(1-p0)$

```python
def gini(samples):
    sample_sum = sum(samples)
    score = 0
    probas = []
    for sample in samples:
        proba = sample / sample_sum
        probas.append(proba)
    for proba in probas:
        score += proba * (1 - proba)
    return score
```

```python
# Проверим
samples = [100, 75, 125]
gini(samples)
```
